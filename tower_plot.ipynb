```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.integrate import cumtrapz

%matplotlib inline
```

# Initial setup

```python
initial_height = 30.0 # rider starts at an initial height of 30 meters.
```

```python
path = 'droptower_vdata.txt'
velocity_data = np.loadtxt(path)
```

# Calculation

```python
t = velocity_data[:, 0]
t_shorter = t[:-1]
```

```python
velocity = velocity_data[:, 1]
acceleration = np.diff(velocity)
position = cumtrapz(velocity, initial = initial_height)
acceleration_positive = acceleration[acceleration>0]
aver_pos = np.mean(acceleration_positive)
```

```python
print('average of positive accelerations:', aver_pos)
```

# Individual plots

```python
plt.plot(t, velocity, 'r+-')
plt.xlabel('time / s')
plt.ylabel('velocity / m*s^-1')
plt.title('velocity of drop tower over time')
plt.show()
```

```python
plt.plot(t_shorter, acceleration, 'g*-')
plt.xlabel('time / s')
plt.ylabel('acceleration / m*s^-2')
plt.title('acceleration of drop tower over time')
plt.show()
```

```python
plt.plot(t, position, 'bo-')
plt.xlabel('time / s')
plt.ylabel('position / m')
plt.title('position of drop tower over time')
plt.show()
```

# 3 lines on the same pair of axes

```python
plt.plot(t, velocity, 'r+-', label = 'velocity')
plt.plot(t_shorter, acceleration, 'g*-', label = 'acceleration')
plt.plot(t, position, 'bo-', label = 'position')
plt.xlabel('time / s')
plt.legend(numpoints=3, loc='best',fontsize='10',frameon=True,shadow=True)
plt.show()
```

# 3 subplots in the same figure

```python
f, axarr = plt.subplots(3, sharex=True)


axarr[0].plot(t, velocity, 'r+-', label = 'velocity')
axarr[0].set_title('velocity (m / s)')
axarr[1].plot(t_shorter, acceleration, 'g*-', label = 'acceleration')
axarr[1].set_title('acceleration (m / s^2)')
axarr[2].plot(t, position, 'bo-', label = 'position')
axarr[2].set_title('position (m)')

axarr[0].set_xlim([0,10])
plt.xlabel('time / s')
plt.tight_layout()
plt.suptitle('sharing the x axis', fontsize=20) #supertitle
plt.subplots_adjust(top=0.85)

plt.show()
```
